we extract the job title, location, date, link and job description


```python
jb_num = 0
while True:
    
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data,'html.parser')
    jobs = soup.find_all('p',{'class':'result-info'})
    
    for job in jobs:
        titre = job.find('a',{'class':'result-title'}).text
        location_tag = job.find('span',{'class':'result-hood'})
        loc = location_tag.text[2:-1] if location_tag else "N/A"
        date = job.find('time', {'class': 'result-date'}).text
        lnk = job.find('a', {'class': 'result-title'}).get('href')
        r2 = requests.get(lnk)
        job_data = r2.text
        job_soup = BeautifulSoup(job_data, 'html.parser')
        job_desc = job_soup.find('section',{'id':'postingbody'}).text
        job_attributes_tag = job_soup.find('p',{'class':'attrgroup'})
        job_atr = job_attributes_tag.text if job_attributes_tag else "N/A"
        jb_num+=1
        print('Job Title:', titre, '\nLocation:', loc, '\nDate:', date, '\nLink:', lnk,"\n", job_atr, '\nJob Description:', job_desc,'\n---')
    url_tag = soup.find('a',{'title':'next page'})
    if url_tag.get('href'):
        url= 'https://boston.craigslist.org' + url_tag.get('href')
        print(url)
    else:
        break
        
print("Total Jobs:", jb_num)

```

# **Create DatarFame to put on it all Data**

```python
 import pandas as pd
```

```python
np_jobs = {}
jb_num = 0
while True:
    
    rp = requests.get(url)
    data = rp.text
    soup = BeautifulSoup(data,'html.parser')
    jobs = soup.find_all('p',{'class':'result-info'})
    
    for job in jobs:
        
        titre = job.find('a',{'class':'result-title'}).text
        location_tag = job.find('span',{'class':'result-hood'})
        loc = location_tag.text[2:-1] if location_tag else "N/A"
        date = job.find('time', {'class': 'result-date'}).text
        lnk = job.find('a', {'class': 'result-title'}).get('href')
        
        job_response = requests.get(lnk)
        job_data = job_response.text
        job_soup = BeautifulSoup(job_data, 'html.parser')
        job_desc = job_soup.find('section',{'id':'postingbody'}).text
        job_attributes_tag = job_soup.find('p',{'class':'attrgroup'})
        job_atr = job_attributes_tag.text if job_attributes_tag else "N/A"
        
        jb_num+=1
        np_jobs[jb_num] = [titre, loc, date, lnk, job_atr, job_desc]
        
        
    tg = soup.find('a',{'title':'next page'})
    if tg.get('href'):
        url= 'https://boston.craigslist.org' + url_tag.get('href')
        print(url)
    else:
        break
        
print("Total Jobs:", jb_num)


```

```python

df = pd.DataFrame.from_dict(np_jobs, orient = 'index', columns = ['Job Title','Location','Date', 'Link', 'Job Attributes', 'Job Description'])
df.head()
```
